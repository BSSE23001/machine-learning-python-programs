### train_test_split
- This functions signature is as follows:
    ```
    train_test_split(
        *arrays,
        test_size=None,
        train_size=None,
        random_state=None,
        shuffle=True,
        stratify=None,
    )
    ```
- This function splits every numpy array we passed to it into two parts so in the following declaration:
  -- The first numpy array (actually a matrix) is splitted and stored into Variables: X_train and X_test
  -- In the Similar way the 2nd numpy array (1d array) is splitted and stored into Variable: y_train and y_test
- The test_size determines how much data we want as test data and ranges from 0 to 1 (probabalistic)
- The train_size is the similar way as test_size.
- random_state => defines the **randomnes of data upon multiple calls (NOT THE SUFFLING ITSELF)**
- shuffle => default true but can be none. If set **none then stratify must also be none**
- stratify => Normally data is not splitted evenly. So to do so we set stratify to our labels array which determines corresponding specie data.

````python
# Extracting data without stratifying with default composition 75% train and 25% test
# The convension Capital X refers to the Matrix
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
````

````python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import mglearn
````

## What is this doing?
This code performs two main operations:
- Creates a Pandas DataFrame from the training data (X_train) using the column names provided by iris_dataset.feature_names.
- Generates a matrix of scatter plots to visualize pairwise relationships between features in the dataset using pd.plotting.scatter_matrix.
- This is especially helpful in exploratory data analysis (EDA) to detect patterns, trends, clusters, or anomalies in multidimensional datasets.

## Step-by-Step Breakdown
### Step 1: Creating a DataFrame
```
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```
- pd.DataFrame(data, columns=None): Constructs a 2D labeled data structure (DataFrame) from an array-like data.
- Parameters:
    - data: array-like, dict, or DataFrame – in this case, X_train.
    - columns: list – names of columns (features) to use for the DataFrame.
- This lets us visualize the training data with readable column labels like sepal length (cm), petal width (cm), etc.

### Step 2: Creating the Scatter Matrix
```
pd.plotting.scatter_matrix(...)
```
#### Function Prototype:
```
pandas.plotting.scatter_matrix(
    frame,
    alpha=0.5,
    figsize=None,
    ax=None,
    grid=False,
    diagonal='hist',
    marker='o',
    density_kwds=None,
    hist_kwds=None,
    range_padding=0.05,
    **kwargs
)
```
#### Parameters Explained:
| Parameter       | Description                                                                         |
| --------------- | ----------------------------------------------------------------------------------- |
| `frame`         | The DataFrame containing numerical values for plotting.                             |
| `alpha`         | Transparency level of the points (`0` = transparent, `1` = opaque).                 |
| `figsize`       | Tuple specifying the overall figure size (width, height in inches).                 |
| `ax`            | Matplotlib Axes object to draw the plots on.                                        |
| `grid`          | If `True`, grid lines are shown in the plots.                                       |
| `diagonal`      | Controls the plot on the diagonal: `'hist'` or `'kde'`.                             |
| `marker`        | Marker style for scatter plots (e.g., `'.'`, `'o'`, `'x'`, etc.).                   |
| `density_kwds`  | Dictionary of keyword arguments for kernel density estimation plots.                |
| `hist_kwds`     | Dict with args passed to `plt.hist()` for histograms (e.g., `bins`).                |
| `range_padding` | Padding added to the range of axes.                                                 |
| `**kwargs`      | Additional keyword args passed to each `plt.scatter()` call, like color, size, etc. |

#### Additional Arguments in Our Code:
| Argument                 | Explanation                                                                                                         |
| ------------------------ | ------------------------------------------------------------------------------------------------------------------- |
| `c=y_train`              | Specifies the color of each point based on the target label (class). Each class will be shown in a different color. |
| `marker='.'`             | Uses dot marker for scatter plots.                                                                                  |
| `hist_kwds={'bins': 20}` | Histogram on the diagonal will use 20 bins.                                                                         |
| `s=60`                   | Size of the markers in scatter plots.                                                                               |
| `alpha=0.8`              | Opacity of the points (80% opaque).                                                                                 |
| `cmap=mglearn.cm3`       | Color map used to color points based on class. `mglearn.cm3` is a predefined colormap from the `mglearn` library.   |

### Why This Produces Multiple Graphs?
- The scatter matrix plots every feature against every other feature in the dataset.
- If there are n features, the scatter matrix will be an n x n grid.
- For the Iris dataset, there are 4 features:
    ```
    Sepal Length | Sepal Width | Petal Length | Petal Width
    ```
- Hence, this will produce a 4×4 = 16 plot matrix:
    - Off-diagonal: scatter plots between every pair of features.
    - Diagonal: histogram (or KDE) of each individual feature.
- This creates a complete visualization to understand how the features relate to each other, and whether classes are separable visually.

### Example Output (Conceptual)
|               | Feature 1        | Feature 2        | Feature 3        | Feature 4        |
| ------------- | ---------------- | ---------------- | ---------------- | ---------------- |
| **Feature 1** | Hist(F1)         | Scatter F1 vs F2 | Scatter F1 vs F3 | Scatter F1 vs F4 |
| **Feature 2** | Scatter F2 vs F1 | Hist(F2)         | Scatter F2 vs F3 | Scatter F2 vs F4 |
| **Feature 3** | Scatter F3 vs F1 | Scatter F3 vs F2 | Hist(F3)         | Scatter F3 vs F4 |
| **Feature 4** | Scatter F4 vs F1 | Scatter F4 vs F2 | Scatter F4 vs F3 | Hist(F4)         |



````python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15),
                           marker='.', hist_kwds={'bins':20}, s=60,
                           alpha=.8, cmap=mglearn.cm3)
````

## K-Nearest Neighbour
- Basic Working of this model is as follows:
    - To make a prediction for a new data point, the algorithm finds the point in the training set that is closest to the new point.
    - Then it assigns the label of this training point to the new data point.
    - The k in k-nearest neighbors signifies that instead of using only the closest neighbor to the new data point, we can consider any fixed number k of neighbors in the training (for example, the closest three or five neighbors).
    - Then, we can make a prediction using the majority class among these neighbors.

**All machine learning models in scikit-learn are implemented in their own classes, which are called Estimator classes.**
### Making Object of Model Class
- The k-nearest neighbors classification algorithm is implemented in the KNeighborsClassifier class in the neighbors module.
- Before we can use the model, we need to instantiate the class into an object.
- This is when we will set any parameters of the model. The most important parameter of KNeighborsClassifier is the number of neighbors, which we will set to 1

````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
````

### knn
- The knn object encapsulates the algorithm that will be used to build the model from the training data, as well the algorithm to make predictions on new data points.
- It will also hold the information that the algorithm has extracted from the training data.
- **In the case of KNeighborsClassifier, it will just store the training set. Because this model's prediction algorithm works by seeing the 'k' number of nearest neighbour and then fits the new data among those labeled neighbours**

#### fit Method
- To build the model on the training set, we call the fit method of the knn object, which takes as arguments the NumPy array X_train containing the training data and the NumPy array y_train of the corresponding training labels
- The fit method returns the knn object itself (and modifies it in place)

**Most models in scikit-learn have many parameters, but the majority of them are either speed optimizations or for very special use cases. We don’t have to worry about the other parameters shown in this representation.**

````python
knn.fit(X_train,y_train)
````

## Making Predictions
- We can put the data for which we want to make prediction into a NumPy 2d array, the number of samples multiplied by the number of features like 1x4 means the array contains 1 row and 4 columns.
- We do this because **scikit-learn always expects two-dimensional arrays for the data.**
- To make a prediction we call **predict** method of the knn object.

````python
# As we know the feature names are of the form and of order as follows:
# Feature Names: ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# So we need to make a data with one row (means one sample in consideration) containing all these 4 features in the same order in column
# Below we are creating one row means we have one sample in consideration and that sample has all 4 essential features in the above order means
# sepal-length = 5, sepal-width = 2.9, petal-length = 1, petal-width = 0.2

X_new = np.array(
    [
    [5,2.9,1,0.2]
    ]
    )
print("The shape of our new data for prediction is {}".format(X_new.shape))
````

````python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted Target Name: {}".format(iris_dataset['target_names'][prediction]))
````

## Evaluating the model
- Our model predicts that this new iris belongs to the class 0, meaning its species is setosa.
- But as for now we can't trust our model that it is predicting the correct species.
- We first have to evaluate our model that how much precise it predicts for any data.
- For evaluation, that data comes in hand which we have splitted into test data during data splitting. Because for the test data we know what the species actually is and we can compare that actual specie with the prediction of our model and calculate the percentile score for our model.
- We can measure how well the model works by computing the accuracy, which is the fraction of flowers for which the right species was predicted.
- We can use the score method of the knn object, which will compute the test set accuracy for us without any required pre-computings.
- Or use the mean function of np module which gives the probability value but require precomputing of the prediction values.

````python
# This score function doesn't need any pre-calculation instead do all the required things itself
print("Test Set Score is: {:.2f}".format(knn.score(X_test,y_test)))
````

````python
# First Calculalting the predictions by our model for the complete testing dataset
y_pred = knn.predict(X_test)
print("The Test Predictions are: {}\n".format(y_pred))
# Now Calculating the Test Score using mean functionality which is calculating the mean of corrected predictions.
print("Test Set Score is: {:.2f}".format(np.mean(y_pred == y_test)))
````
